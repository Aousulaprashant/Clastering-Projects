```python
import pandas as pd
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import numpy as np
```

```python
iris_data=load_iris()
```

```python
iris_data
```

```python
iris_data=pd.DataFrame(data=iris_data.data,columns=iris_data.feature_names)
```

```python
iris_data.head()
```

```python
iris_data.shape
```

```python
from sklearn.decomposition import PCA

pca=PCA(n_components=2)
Reduced_data=pca.fit_transform(iris_data)
```

```python
Reduced_data.shape
```

```python
plt.scatter(Reduced_data[:,0],Reduced_data[:,1])
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test=train_test_split(Reduced_data,test_size=0.33,random_state=22)
```

```python
X_train.shape,X_test.shape
```

```python
from sklearn.preprocessing import StandardScaler

standered_data=StandardScaler()

standered_data.fit_transform(X_train)
```

```python
print(X_train)
```

```python
import scipy.cluster.hierarchy as sc

plt.figure(figsize=(20,7))
plt.title("Dendograms")

sc.dendrogram(sc.linkage(X_train,method='ward'))
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster=AgglomerativeClustering(n_clusters=2,linkage='ward')
cluster.fit(X_train)
```

```python
cluster.labels_
```

```python
plt.scatter(X_train[:,0],X_train[:,1],c=cluster.labels_)
```
